```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Latency Section

```python
region = [0, np.log2(2**15), np.log2(2**15), np.log2(2**18), np.log2(2**18), np.log2(2**21)]
```

```python
region
```

```python
latencies = [2, 2, 12, 12, 81, 81]
```

```python
plt.xlabel('log2(memory_region)')
plt.ylabel('Latency (ns)')
plt.title('Memory Latencies for different memory regions')
plt.text(4, 6, 'L1Cache=2**15 bytes', fontsize = 10)
plt.text(15.5, 7.5, 'L2Cache=2**18 bytes', fontsize = 10)
plt.text(18, 75, 'Memory=2**21 bytes', fontsize = 10)
plt.plot(region, latencies)
```

### File Cache Size Section

```python
file_sizes = ["128 MB", "256 MB", "512 MB", "1 GB", "2 GB", "8 GB", "16 GB", "32 GB"]
```

```python
cache_ios = [6.8, 5.4, 6.1, 6.7, 24, 2009, 4930, 5537]
```

```python
plt.xlabel('Size of File in Bytes')
plt.ylabel('Read I/O in microseconds')
plt.title('Read I/Os for different file sizes')
plt.plot(file_sizes, cache_ios)
```
